```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import skew, kurtosis
```

```python
def descargar_datos_yf(tickers, start, end):
    """
    Descarga precios de cierre de Yahoo Finance.
    """
    data = yf.download(tickers, start=start, end=end)['Close']
    return data
```

```python
def calcula_retornos(df):
    """
    Calcula retornos porcentuales.
    """
    retornos = df.pct_change().dropna()
    return retornos
```

```python
def resumen_estadistico_retornos(df):
    """
    Calcula medidas de tendencia central, dispersión y forma.
    """
    stats = pd.DataFrame({
        'Media': df.mean(),
        'Volatilidad': df.std(),
        'Asimetría': df.apply(skew),
        'Kurtosis': df.apply(kurtosis),
        'Máximo': df.max(),
        'Mínimo': df.min()
    })
    return stats
```

```python
def plot_retornos(df):
    """
    Visualiza retornos con serie de tiempo, histograma y boxplot.
    """
    fig, axes = plt.subplots(3, 1, figsize=(12, 14))
    
    # Serie de tiempo
    df.plot(ax=axes[0])
    axes[0].set_title("Series de tiempo de retornos")
    
    # Histograma
    df.hist(ax=axes[1], bins=50)
    axes[1].set_title("Histograma de retornos")
    
    # Boxplot
    sns.boxplot(data=df, ax=axes[2])
    axes[2].set_title("Boxplot de retornos")
    
    plt.tight_layout()
    plt.show()
```

```python
tickers = ['AAPL', 'JPM', 'TSLA']
periodo = ('2021-01-01', '2024-01-01')

# Descargar precios
precios = descargar_datos_yf(tickers, periodo[0], periodo[1])

# Calcular retornos
retornos = calcula_retornos(precios)

# Resumen estadístico
stats = resumen_estadistico_retornos(retornos)
print(stats)

# Visualización
plot_retornos(retornos)
```
